```
# Setting Database
import os

path ='/content/drive/My Drive/Mini Project/images/richard mille watch'

for filename in os.listdir(path):
    i =cv2.imread(os.path.join(path,filename))
    i =cv2.cvtColor(i,cv2.COLOR_BGR2RGB)
    print(i.shape)
```

```
from skimage.io import imread
from skimage.transform import resize
import numpy as np
```

```
target =[]  #list of target
flat_data=[] #flattened image data
images = []

path = '/content/drive/My Drive/Mini Project/images'

for dirpath,dirnames,filenames in os.walk(path):
    for filename in filenames:
        
        category = dirpath.split('/')[-1]
        
        target.append(category)
        img =imread(f'{path}/{category}/{filename}')
        img_resize = resize(img,(50,50,3))
        flat_data.append(img_resize.flatten())
        images.append(img_resize)
    
        
flat_data = np.array(flat_data)
target = np.array(target)
images = np.array(images)

print(flat_data.shape)
print(target)
```

```
# Converting Data Set into dataframe
import pandas as pd

df =pd.DataFrame(flat_data)

df['Target']=target
```

```
df.head()
```

```
x=flat_data
y=df.Target
```

```
# Converting into training data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y)

x_test.shape
```

```
from sklearn.svm import SVC
model=SVC()
model.fit(x_train,y_train)
```

```
y_pred=model.predict(x_test)
y_pred
```

```
# showing model accuracy
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```
# Using KNN to improve accuracy
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=6)
model.fit(x_train,y_train)
pred=model.predict(x_test)
pred
```

```
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,pred))
print(confusion_matrix(y_test,pred))
```
